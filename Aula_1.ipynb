```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/49 Data Visualization: criando gráficos com bibliotecas Python/imigrantes_canada.csv')
```

```python
df
```

```python
df.info()
```

```python
df.set_index('País', inplace=True)
# mudar índice para os nomes dos países
```

```python
anos = list(map(str, range(1980, 2014)))
# atribuir à variável anos uma lista de strings no intervalo entre os os anos de 1980 e 2013
```

```python
anos
```

```python
brasil = df.loc['Brasil', anos]
# selecionar os números de imigrantes do Brasil dentro do intervalo de anos especificado
```

```python
brasil
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
# criar dicionário com a coluna ano sendo os índices da serie da célula anterior e a coluna
# imigrantes sendo os valores da serie da célula anterior
# convertendo tanto os índices quanto os valores em lista
dados_brasil = pd.DataFrame(brasil_dict)
# atribuindo à variável dados_brasil a criação de dataframe com base no dicionário brasil_dict
```

```python
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
# modificando o tamanho do gráfico
plt.figure(figsize=(8, 4))
# criando gráfico de linhas
  # eixo x = horizontal = abscissas
  # eixo y = vertical = ordenadas
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
# criando um título
plt.title('Imigração do Brasil para o Canadá')
# criando um rótulo para o eixo x
plt.xlabel('Ano')
# criando um rótulo para o eixo y
plt.ylabel('Número de imigrantes')
# alterando os valores do eixo x
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
# retirando as informações técnicas sobre o gráfico, mostrando apenas este
plt.show()
```

```python
america_sul = df.query('Região == "América do Sul"')
```

```python
america_sul_sorted = america_sul.sort_values('Total', ascending=True)
```

```python
cores = []
for pais in america_sul_sorted.index:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('silver')

fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canadá no período de 1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i, v in enumerate(america_sul_sorted['Total']):
  ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)

fig.savefig('imigracao_america_sul.png', transparent=False, dpi=300, bbox_inches='tight')

plt.show()
```

```python
# PARA CRIAR GRÁFICOS ANIMADOS

## obter o DataFrame apenas com os dados do Brasil
## mudar o tipo de dados da coluna que contém os anos para int ao invés de manter como strings
dados_brasil['ano'] = dados_brasil['ano'].astype(int)
## criar um bloco de código para construir a animação, importando plotly.graph_objs, um módulo
## da biblioteca Plotly que contém classes para criar visualizações de dados interativas e
## personalizadas
import plotly.graph_objs as go
## criar figura vazia
fig = go.Figure()
## adicionar linha ao gráfico com o argumento Scatter
## iloc[0] para que o gráfico seja exibido sem linha antes de clicar no botão play
fig.add_trace(
  go.Scatter(x=[dados_brasil['ano'].iloc[0]], y=[dados_brasil['imigrantes'].iloc[0]],
             mode='lines', name='Imigrantes', line=dict(width=4))
)
## definir o título do gráfico e as configurações dos eixos x e y
  ### text='<b>Imigração do Brasil para o Canadá no período de 1980 a 2013</b>': define o texto
  ### do título como uma string formatada em negrito (usando as tags HTML "<b>" e "</b>")
  ### x=0.12: define a posição horizontal do título no layout, em relação à largura da figura.
  ### O valor 0.12 especifica que o título começará a 12% da largura da figura
  ### xanchor='left': define o alinhamento horizontal do título. O valor 'left' significa que o
  ### título será alinhado à esquerda do layout
  ### font=dict(size=20): define o tamanho do texto do título
  ### xaxis e yaxis -> dict
    #### range
    #### autorange=False: define se os limites do eixo serão ajustados automaticamente (True)
    #### ou não (False)
    #### title='<b>Ano</b>': define o título do eixo. Neste caso, o eixo x terá o título "Ano",
    #### que é formatado em negrito (usando as tags HTML "<b>" e "</b>")
    #### title='<b>Número de imigrantes</b>': define o título do eixo. Neste caso, o eixo y terá
    #### o título "Número de imigrantes", que é formatado em negrito (usando as tags HTML "<b>"
    #### e "</b>")
fig.update_layout(
    title=dict(
        text='<b>Imigração do Brasil para o Canadá no período de 1980 a 2013</b>',
        x=0.12,
        xanchor='left',
        font=dict(size=20)
    ),
    xaxis=dict(range=[1980, 2013], autorange=False, title='<b>Ano</b>'),
    yaxis=dict(range=[0, 3000], autorange=False, title='<b>Número de imigrantes</b>'),
    ## adicionar o botão Play com o argumento updatemenus (lista que define as opções de menu para
    ## a figura); seus valores são dict
      ### type='buttons': define que o menu será composto por botões
      ### showactive=False: define que nenhum botão estará ativo inicialmente
      ### buttons=[dict(label='Play', method='animate', args=[None, {'frame': {'duration': 100,
      ### 'redraw': True}, 'fromcurrent': True}])]: define o botão que será exibido no menu
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
        )]
    )],
    ## largura e altura do gráfico
    width=1000,
    height=500
)
## definir as configurações de animação
  ### frames -> lista de objetos Frame do Plotly, com informações dos dados para cada quadro da
  ### animação
  ### for para criar um objeto Frame para cada ano no conjunto de dados, até o último ano
  ### iloc para selecionar os valores do DataFrame dados_brasil a partir do índice 0 até o
  ### índice atual i+1
  ### a cada iteração, a lista frames é preenchida com um novo objeto Frame
  ### a lista frames é atribuída à propriedade frames do objeto Figure (fig) (permite que a
  ### animação seja exibida no gráfico quando o botão de controle de animação é pressionado; cada
  ### frame contém os dados do gráfico para um determinado ano e é exibido em sequência quando
  ### o botão "Play" é clicado)
frames = [
    go.Frame(
        data=[
            go.Scatter(
                x=dados_brasil['ano'].iloc[:i+1],
                y=dados_brasil['imigrantes'].iloc[:i+1])]) for i in range(len(dados_brasil))
]
fig.frames = frames
fig.show()
```
